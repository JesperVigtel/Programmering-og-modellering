# Ledeegenskaper til fingre 

Tilfeldig uvalg av ofre fra elever p√• Valler vgs

```python
from pylab import *

data = loadtxt("Ledefinger.csv", skiprows = 1, delimiter = ";")


tommel = data[:,0]
peke = data[:,1]


a ,b ,c,d,e = polyfit(peke, tommel,4)


def modell(x):
    return a*x**4+b*x**3+c*x**2+d*x+e

x = linspace(5,19,1000)
y = modell(x)



t = mean(tommel)
p = mean(peke)

print(p,t)

plot(peke,tommel,linestyle = "--", marker = ".", label = "Datapunkter")
plot(x,y, label = "Tilpasset kurve")
title("Ledevene til fingre")
xlabel("Ledeevne til pekefinger")
ylabel("Ledeevne til tommelfinger")
legend(["M√•linger","Tilpasset"])
show()
```

M√•lingene sier noe om sammenhengen mellom ledeevnen til pekefingeren og tommelfingeren til eleven. Ut ifra grafen ser det ut som at det er en slags sammenheng, men at det trengs flere m√•linger for √• bekrefte. Den oransje grafen er en tilapasset graf som viser et eksempel p√• en tydeligere sammenheng. Men igjen, s√• trengs det flere m√•linger for √• f√• den mer n√∏yaktig.




## Gjennomsnitt og standaravvik



```python
def gjennomsnitt(verdier):
    n = len(verdier)
    summen = 0
    for i in range(n):# summen = sum(verdier) (litt juks)
        summen = summen + verdier[i]
    snitt = summen/n
    return snitt
def standaravvik(verdier):
    n = len(verdier)
    snitt = gjennomsnitt(verdier)
    avvik = 0
    for verdi in verdier:
        avvik = avvik + (verdi - snitt)**2
    avvik = (avvik/n)**0.5
    return avvik
    
    
print("Gjennomsnittet av ledeevnen til tommelfingeren er: ", gjennomsnitt(tommel), "og til pekefingeren: ", gjennomsnitt(peke))
print("Standaravviket til tommelfinger er: ", standaravvik(tommel), "og til pekefinger: ", standaravvik(peke))

```

Vi har kommet frem til at det er ganske tilfeldig hvordan ledeegenskapene til fingere er, og at vi tror det har noe √• si med hvor svett man er p√• fingeren. Vi fikk ikke tatt nok tester til √• bekrefte dettte, men mistenker at det er en sammenheng mellom ledeegenskaper til tommel og pekefinger.
